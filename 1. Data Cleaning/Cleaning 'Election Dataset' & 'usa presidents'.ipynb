```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Data - Wikipedia (https://en.wikipedia.org/wiki/Voter_turnout_in_United_States_presidential_elections)
election_data=pd.read_csv('/content/drive/MyDrive/CP project/Project/Raw Datasets/Election dataset.csv',sep=';',encoding='ISO-8859-1',header=[0, 1])
election_data
```

```python
# Replacing '' as the dataset was extrapolated from a Wikipedia page
election_data.replace('', np.NaN, inplace=True)
```

```python
# Data examination
election_data.columns = ['Year',
              'Voting-age Population (VAP)',
              'Voting-eligible Population (VEP)',
              'Turnout',
              'VAP Turnout',
              'VEP Turnout']

election_data['Year'] = election_data['Year'].replace('2024* (estimated)', '2024')

election_data.columns
```

```python
election_data.info()
# we need to confert some variables into numeric ones
```

```python
election_data.describe()
```

```python
# missing data
election_data.isnull().sum()
```

```python
# Dropping VEP data as we do not anough records
election_data.drop(['Voting-eligible Population (VEP)','VEP Turnout'],axis=1,inplace=True)
```

```python
election_data['Voting-age Population (VAP)'] = election_data['Voting-age Population (VAP)'].replace({',': ''}, regex=True)
election_data['Turnout'] = election_data['Turnout'].replace({',': ''}, regex=True)
election_data['VAP Turnout'] = election_data['VAP Turnout'].replace({',':'','%': ''}, regex=True)

election_data['Year'] = election_data['Year'].astype(int)
election_data['Voting-age Population (VAP)'] = pd.to_numeric(election_data['Voting-age Population (VAP)'])
election_data['Turnout'] = pd.to_numeric(election_data['Turnout'])
election_data['VAP Turnout'] = pd.to_numeric(election_data['VAP Turnout'])
```

```python
# Let's add another wikipedia dataset containing the presidents, their mandate and party
president_data=pd.read_csv('/content/drive/MyDrive/CP project/Project/Raw Datasets/usa presidents.csv',sep='|')
president_data.head()
```

```python
president_data.columns
```

```python
president_data.columns = president_data.columns.str.strip()
```

```python
president_data.info()
```

```python
# In order to do the merging, both variables need to be int, but the president_dataset
# contains strange values inside 'Election'
non_numeric_rows=president_data[president_data['Election'].str.contains('[a-zA-Z]')]
display(non_numeric_rows)
```

```python
revised_presidents = {
    'John Tyler': 1840,
    'Millard Fillmore': 1850,
    'Andrew Johnson': 1864,
    'Chester A. Arthur': 1880,
    'Grover Cleveland': 1892,
    'Gerald Ford': 1974
}

# Use the dictionary to update the 'Election' column for the specified presidents
for president, year in revised_presidents.items():
    president_data.loc[president_data['President'].str.strip() == president, 'Election'] = year

president_data
```

```python
president_data['Election'] = president_data['Election'].astype(int)
print(president_data['Election'].isnull().sum())
print(election_data['Year'].dtypes)
print(president_data['Election'].dtypes)
```

```python
# Renaming and Converting the 'Election' column, and then merging the 2 datasets
president_data.rename(columns={'Election': 'Year'}, inplace=True)
election_data = pd.merge(election_data, president_data, on='Year', how='left')
election_data.head()
```

```python
# Let's add the 2024 president
data_2024 = {
    'Year': 2024,
    'President': 'Donald Trump',
    'Party': 'Republican',
    'Mandate Years': 'TBD'}

# update only the variables defined in data_2024
for column, value in data_2024.items():
    election_data.at[election_data.index[-1], column] = value

election_data.set_index('Year',inplace=True)
election_data
```

```python
votes_by_party=pd.read_csv('/content/drive/MyDrive/CP project/Project/Cleaned Datasets/Votes by party 1976-2020.csv')
votes_by_party.head()
```

```python
# Merging
election_data = pd.merge(election_data, votes_by_party, on='Year', how='left')
election_data.rename(columns={'DEMOCRAT':'Democratic',
                              'REPUBLICAN':'Republican',
                              'OTHER':'Other',
                             'LIBERTARIAN':'Libertarian'},inplace=True)

election_data_sorted = election_data.sort_values(by='Year',ascending=True)



display(election_data)
```

```python
election_data.rename(columns={'Party':"President's Party"},inplace=True)
election_data["President's Party"] = election_data["President's Party"].str.strip()
```
