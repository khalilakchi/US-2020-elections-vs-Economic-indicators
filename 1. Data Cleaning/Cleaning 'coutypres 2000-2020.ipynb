```python
county_data.columns
county_data.columns = county_data.columns.str.strip()
```

```python
county_data.rename(columns={'year':'Year',
                    'state_po':'State_initials',
                    'state':'State',
                    'county_name':'County'},inplace=True)
```

```python
pivoted_data = county_data.pivot_table(
    index=['Year','State_initials','State', 'County'],  # Group by year, state, and county
    columns='party',                        # Separate the votes by party
    values='candidatevotes',                 # Aggregate the votes for each party
    aggfunc='sum',                          # Sum the votes (in case there are duplicates)
    fill_value=0                             # Replace NaN with 0 for counties with no votes for a party
)

print(pivoted_data.columns)

pivoted_data.rename(columns={'Democrat': 'Democratic'}, inplace=True)
pivoted_data.columns = [col.title() for col in pivoted_data.columns]




pivoted_data
```
