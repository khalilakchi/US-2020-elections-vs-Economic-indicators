```python
# source: https://www.bea.gov/
gdp_data=pd.read_csv('/content/drive/MyDrive/CP project/Project/Raw Datasets/GDP data.csv',skiprows=3,sep=';')
gdp_data
```

```python
gdp_data.columns
```

```python
print(gdp_data.isnull().sum())
gdp_data[gdp_data[['GeoName','2020']].isna()]
```

```python
# let's immediately drop the footnotes and the column 'Unnamed:5'
gdp_data = gdp_data.dropna(subset=['GeoName'])
gdp_data.drop(['Unnamed: 5','GeoFips','LineCode'],axis=1,inplace=True)
gdp_data.tail()
```

```python
gdp_data['Description'].unique()
```

```python
comma_count = gdp_data['GeoName'].str.count(',')

# Extract rows where the number of commas is more than 1
rows_with_multiple_commas = gdp_data[comma_count > 1]

# Display the rows with more than one comma in 'GeoName'
print(rows_with_multiple_commas)
```

```python
gdp_data[['County', 'State']] = gdp_data['GeoName'].str.rsplit(',', n=1, expand=True)
# str.rssplit(',', 1, expand=True) is used to split the GeoName column into two parts based on the comma.
#',': The string is split at the comma ,.
# rsplit is used to ensure that we split at the last comma, so any commas in the county name (like "Buena Vista + Lexington") won’t interfere with the split.
# 1: This means split only once. It tells pandas to only split the string at the first comma it encounters. If there are multiple commas in the string (like in "Augusta, Staunton + Waynesboro, VA"), pandas will split only at the first comma and leave the rest of the string intact.
# expand=True: This tells pandas to expand the result into separate columns. Without this, you would get a list or a series, but with expand=True, each part after the split will be placed into its own column.


# Clean the 'State' column to remove any trailing characters (like '*' or spaces)
gdp_data['State'] = gdp_data['State'].str.replace(r'\*|\s+', '', regex=True)

# 2. Strip any leading/trailing whitespace from the 'County' and 'State' columns
gdp_data['County'] = gdp_data['County'].str.strip()
gdp_data['State'] = gdp_data['State'].str.strip()

# 3. Display the result
print(gdp_data[['County', 'State', 'GeoName']].head())
```

```python
gdp_data['State'].unique()
```

```python
gdp_data.drop('GeoName',axis=1,inplace=True)
gdp_data[gdp_data['State']=='VA']
```

```python
gdp_data = gdp_data.pivot_table(index=['County', 'State'], columns='Description', values='2020', aggfunc='first')
# This means that the values in the Description column will be turned into new columns in the pivoted DataFrame. So, each unique value in the 'Description' column will become its own column.
#This specifies that the data in the 2020 column will be used as the values for the new columns (created by Description). These values will be placed into the pivoted DataFrame, organized by the new rows and columns.
#aggfunc='first':This argument specifies the aggregation function to use when there are multiple values for a given combination of 'County', 'State', and 'Description'. The 'first' function means that, if there are multiple values for that combination, only the first value encountered will be kept. Other aggregation options could be sum, mean, etc., depending on the desired behavior.


# Rename the columns for clarity (optional step)
gdp_data.columns.name = None  # Remove the column name (which was 'Description')
gdp_data.rename(columns={
    'Real GDP (thousands of chained 2017 dollars)': 'Real GDP (2017 dollars)',
    'Chain-type quantity indexes for real GDP': 'Chain-type GDP Index',
    'Current-dollar GDP (thousands of current dollars)': 'Current-dollar GDP'
}, inplace=True)
```

```python
gdp_data.head()
```

```python
print(gdp_data.info())

print(gdp_data.isna().sum())
```

```python
# although there are no Nan variables, there are some strange characters that do not allow us
# to convert them into int/floats variables

# Check for 'NM' in any of the three columns
gdp_data[
    gdp_data['Chain-type GDP Index'].str.contains(r'NM|\(NA\)') |
    gdp_data['Current-dollar GDP'].str.contains(r'NM|\(NA\)') |
    gdp_data['Real GDP (2017 dollars)'].str.contains(r'NM|\(NA\)')]

```

```python
gdp_data['Chain-type GDP Index'] = gdp_data['Chain-type GDP Index'].replace({'(NM)': np.nan, '(NA)': np.nan})
gdp_data['Current-dollar GDP'] = gdp_data['Current-dollar GDP'].replace({'(NM)': np.nan, '(NA)': np.nan})
gdp_data['Real GDP (2017 dollars)'] = gdp_data['Real GDP (2017 dollars)'].replace({'(NM)': np.nan, '(NA)': np.nan})

gdp_data.isnull().sum()
```

```python
# Remove commas and strip spaces in order to convert the variables as floats
gdp_data['Chain-type GDP Index'] = (
    gdp_data['Chain-type GDP Index']
    .str.replace(',', '', regex=True)
    .str.strip()
    .astype(float)  # Use float in case of decimal values
)

gdp_data['Current-dollar GDP'] = (
    gdp_data['Current-dollar GDP']
    .str.replace(',', '', regex=True)
    .str.strip()
    .astype(float)
)

gdp_data['Real GDP (2017 dollars)'] = (
    gdp_data['Real GDP (2017 dollars)']
    .str.replace(',', '', regex=True)
    .str.strip()
    .astype(float)
)

gdp_data.info()
```

```python
# Let's replace the Nan values with the means of the state

state_means = gdp_data.groupby('State').mean().round(2)
print(state_means)

gdp_data['Chain-type GDP Index'] = gdp_data.apply(
    lambda row: state_means.loc[row.name[1], 'Chain-type GDP Index'] if pd.isna(row['Chain-type GDP Index']) else row['Chain-type GDP Index'],
    axis=1
)

gdp_data['Current-dollar GDP'] = gdp_data.apply(
    lambda row: state_means.loc[row.name[1], 'Current-dollar GDP'] if pd.isna(row['Current-dollar GDP']) else row['Current-dollar GDP'],
    axis=1
)

gdp_data['Real GDP (2017 dollars)'] = gdp_data.apply(
    lambda row: state_means.loc[row.name[1], 'Real GDP (2017 dollars)'] if pd.isna(row['Real GDP (2017 dollars)']) else row['Real GDP (2017 dollars)'],
    axis=1
)
```
