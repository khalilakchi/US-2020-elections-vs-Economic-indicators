```python
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
poverty_data=pd.read_csv('/content/drive/MyDrive/CP project/Project/Raw Datasets/poverty data.csv',skiprows=3,sep=';')
poverty_data.head()
```

```python
poverty_data.info()
```

```python
poverty_data.replace('.', np.nan, inplace=True)

poverty_data.isnull().sum()
```

```python
# Let's check if FIPS code can be used as County identifiers
# Conta quante volte ogni 'County FIPS Code' appare
fips_counts = poverty_data['County FIPS Code'].value_counts()
print(fips_counts)
```

```python
# This because it's combination between the FIPS and the state to be unique
poverty_data[poverty_data.duplicated(subset=['Postal Code', 'County FIPS Code'], keep=False)]
```

```python
poverty_data['Postal Code'].unique()
```

```python
# the the last columns contain too many Nan values to be analyzed
poverty_data.drop(['Poverty Estimate, Age 0-4','90% CI Lower Bound.7','90% CI Upper Bound.7',
                   'Poverty Percent, Age 0-4','90% CI Lower Bound.8','90% CI Upper Bound.8'
                  ],inplace=True,axis=1)


# Let's now select only the interested columns
poverty_data=poverty_data[['County FIPS Code','Postal Code','Name','Poverty Estimate, All Ages',
                         'Poverty Percent, All Ages','Median Household Income',
                          'Poverty Estimate, Age 0-17','Poverty Percent, Age 0-17'
                         ]]

poverty_data.head()
```

```python
poverty_data['Poverty Estimate, All Ages'] = poverty_data['Poverty Estimate, All Ages'].str.replace(',', '').str.replace('.', '').astype(float)
poverty_data['Poverty Percent, All Ages'] = poverty_data['Poverty Percent, All Ages'].str.replace(',', '.').astype(float)
poverty_data['Median Household Income'] = poverty_data['Median Household Income'].str.replace(',', '').astype(float)
poverty_data['Poverty Estimate, Age 0-17'] = poverty_data['Poverty Estimate, Age 0-17'].str.replace(',', '').str.replace('.', '').astype(float)
poverty_data['Poverty Percent, Age 0-17'] = poverty_data['Poverty Percent, Age 0-17'].str.replace(',', '.').astype(float)

poverty_data.rename(columns={
    'Postal Code':'State',
    'Name':'County',
    'County FIPS Code':'FIPS'
},inplace=True)


poverty_data.info()
```

```python
# Just to avoid repetitions and to make the dataset more consistent with the other datasets
# let's remove the word 'County'
poverty_data['County'] = poverty_data['County'].str.replace(' County', '', regex=False)
```

```python
poverty_data.to_csv('/content/drive/MyDrive/CP project/Project/Cleaned Datasets/2020 poverty data.csv',index=False)
```

```python
poverty_state_data = poverty_data.pivot_table(
    index="State",
    values=[
        "Poverty Estimate, All Ages", "Poverty Percent, All Ages",
        "Median Household Income", "Poverty Estimate, Age 0-17",
        "Poverty Percent, Age 0-17"
    ],
    aggfunc={
        "Poverty Estimate, All Ages": "sum",
        "Poverty Estimate, Age 0-17": "sum",
        "Poverty Percent, All Ages": "mean",
        "Poverty Percent, Age 0-17": "mean",
        "Median Household Income": "mean"
    }
)
poverty_state_data
```
