```python
election_data=pd.read_csv('/content/drive/MyDrive/CP project/Project/Cleaned Datasets/Election Dataset Cleaned.csv')

election_data
```

```python
# Define the mapping dictionary
party_color_map = {
    'Republican': 'Red',
    'Democratic': 'Blue',
    'Libertarian': 'Yellow',
    'Other':'Green'
}

election_data["President's Party"].map(party_color_map)
```

```python
election_data.info()
```

```python
# Extract colors for the winning party based on the color map
president_colors = election_data["President's Party"].map(party_color_map)

fig, ax = plt.subplots(figsize=(14, 8))
bars = ax.bar(election_data['Year'],
              election_data['VAP Turnout'],
              color=president_colors,
              edgecolor='black')

ax.set_title('Turnout by Election Year and Winning Party', fontsize=16)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Turnout', fontsize=12)

ax.set_xticks(election_data['Year'])
ax.set_xticklabels(election_data['Year'], rotation=45)

plt.show()
```
