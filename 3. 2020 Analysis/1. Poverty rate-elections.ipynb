```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
plt.rc('figure', figsize=(10, 5))
```

```python
votes=pd.read_csv('/content/drive/MyDrive/CP project/Project/Cleaned Datasets/Votes Party per state 1976-2020.csv')
votes_2020= votes[votes["Year"]==2020] #select 2020 data
votes_2020['State'] = votes_2020['State'].str.strip().str.title()
votes_2020
votes_2020.to_csv("elections 2020")
```

```python
poverty_data= pd.read_csv('/content/drive/MyDrive/CP project/Project/Cleaned Datasets/2020 poverty data.csv')
poverty_data['County'] = poverty_data['County'].str.strip().str.title()
poverty_data #in counties there are also state data
```

```python
state_names= votes_2020["State"]
state_names_set= set(state_names)

filtered_data = poverty_data[poverty_data['County'].isin(state_names_set)& (poverty_data['FIPS'] == 0)]

filtered_data.rename(columns={"State":"State abb", "County":"State"}, inplace=True)
filtered_data
```

```python
election_poverty =pd.merge(filtered_data, votes_2020, how="right", on="State")
election_poverty
parties=["Democratic",	"Libertarian"	,"Other",	"Republican"]
election_poverty.drop(columns=["FIPS","Poverty Percent, Age 0-17","Poverty Estimate, Age 0-17"], inplace=True)
election_poverty["Winner"]=election_poverty[parties].idxmax(axis=1)
election_poverty
```

```python
states_pres=pd.read_csv('/content/drive/MyDrive/CP project/Project/Raw Datasets/1976-2020-president.csv')
states_pres
filtered_state=states_pres[["state_po","state_fips"]].drop_duplicates()
filtered_state.rename(columns={"state_po":"State abb", "state_fips":"STATEFP"}, inplace=True)
filtered_state
```

```python
election_poverty_fips=pd.merge(filtered_state,election_poverty, how="right", on="State abb")
election_poverty_fips
```

```python
ax=plt.scatter(election_poverty["Poverty Percent, All Ages"],election_poverty["Democratic"], marker="s",color="mediumvioletred",edgecolor="black", s=100)
plt.xlabel("Poverty Percent", fontsize=10, fontweight="bold")
plt.ylabel("Democratic votes", fontsize=10, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.7)
plt.title("Relation Between Poverty Percent and Democratic votes", fontsize=14, fontweight="bold")
plt.show()
```

```python
ax=plt.scatter(election_poverty["Poverty Percent, All Ages"],election_poverty["Republican"], marker="s",color="mediumvioletred",edgecolor="black", s=100)
plt.xlabel("Poverty Percent", fontsize=10, fontweight="bold")
plt.ylabel("Republican votes", fontsize=10, fontweight="bold")
plt.grid(True, linestyle="--", alpha=0.7)
plt.title("Relation Between Poverty Percent and Republican votes", fontsize=14, fontweight="bold")
plt.show()
```

```python
correlation_matrix = election_poverty[['Poverty Percent, All Ages', 'Democratic', 'Republican']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation matrix between poverty % and party vote')
plt.show()
```

```python
sorted_poverty_data= election_poverty.sort_values(by="Median Household Income", ascending=False)
top_5 = sorted_poverty_data.head(5)
bottom_5 = sorted_poverty_data.tail(5)

top_5
```

```python

party_color_map = {
    'Republican': 'red',
    'Democratic': 'blue',
    'Libertarian': 'green',
    'Other': 'grey'
}

election_poverty['colour'] = election_poverty['Winner'].map(party_color_map)
election_poverty

sorted_poverty_data['colour'] = sorted_poverty_data['Winner'].map(party_color_map)
sorted_poverty_data


plt.figure(figsize=(20,10))
plt.bar(sorted_poverty_data['State'], sorted_poverty_data['Median Household Income'],color=sorted_poverty_data["colour"])
plt.title('State for Median Household Income and elected party')
plt.xlabel('State')
plt.xticks(rotation=90)
plt.ylabel('Median Household Income')
plt.show()
```

```python
import plotly.express as px

fig = px.choropleth(election_poverty_fips,
                    locations="State abb",
                    locationmode="USA-states",
                    color="Winner",
                    scope="usa",
                    color_discrete_map=party_color_map,
                    title="USA and Poverty Percent")

for i, row in election_poverty_fips.iterrows():
    fig.add_scattergeo(
        locations=[row['State abb']],
        locationmode="USA-states",
        text=f"{row['Poverty Percent, All Ages']:.2f}%",
        mode="text",
        textfont=dict(size=10, color="black"),
        showlegend=False)

fig.show()
```
