Multivariate quantile regression (MQR) was chosen over simple linear regression (OLS) for its ability to address several critical limitations when analyzing the number of votes for each American party in the 2020 elections.
Unlike simple linear regression, which models only the mean of the dependent variable and assumes homoscedasticity, quantile regression can explore relationships across different quantiles of the vote distribution. This provides a more nuanced understanding of how predictors, influence votes at various levels, particularly in a dataset with log-transformed scaling. The log transformation, while useful for dealing with skewed data, can introduce heteroscedasticity, meaning that the variance of the error terms is not constant. In this context, MQR is particularly valuable because it is robust to heteroscedasticity and can offer more reliable estimates across different quantiles of the vote distribution, rather than just the mean.

In addition to addressing heteroscedasticity, multivariate quantile regression allows for the simultaneous modeling of multiple dependent variables (such as the votes for the Democratic, Green, Libertarian, and Republican parties). This joint modeling captures the correlations between vote counts for different parties and allows for a more comprehensive perspective. By estimating the relationships at various quantiles , MQR provides insights into how economic factors, such as unemployment rate, poverty percentage, and real GDP, influence votes at different points of the vote distribution. For example, it may reveal that the impact of unemployment is stronger for the upper quantiles of Republican votes compared to the lower quantiles, suggesting that economic shifts may affect wealthier voters more intensely.

The choice of multivariate quantile regression is especially important given the political significance of the dataset. It allows for a deeper understanding of how economic factors affect not only the central tendency of votes (as in OLS) but also the more extreme parts of the distribution—both high and low levels of support. This provides a clearer picture of the diverse ways in which economic changes can shape voter behavior across different groups, making MQR better suited to the complexities of analyzing political voting patterns.

In conclusion, by using multivariate quantile regression, this analysis will offer a more robust and detailed view of how economic factors influenced voting behavior during the 2020 U.S. elections. It will allow for examining the conditional relationships at different quantiles, providing a clearer picture of voter behavior, especially in light of the heteroscedasticity and complex nature of the data.

```python
import statsmodels.formula.api as smf

quantile_model = smf.quantreg('DEMOCRAT_log ~ Q("unemp rate(log)") + Q("Poverty Percent, All Ages(log)") + Q("Real GDP(log)")', filtered_df)
quantile_fit = quantile_model.fit(q=0.5)
quantile_fit.summary()
```

Pseudo R squared: 0.7 suggests that the model explain well the variability in democrats votes. REAL gdp (log) is 1.14 and has a positive relation with dem votes.
unemployment rate and poverty percent are significant but with less effects

```python
quantile_model = smf.quantreg('REPUBLICAN_log ~ Q("unemp rate(log)") + Q("Poverty Percent, All Ages(log)") + Q("Real GDP(log)")', filtered_df)
quantile_fit = quantile_model.fit(q=0.5)
quantile_fit.summary()
```

the model is less strong than the democrats one: the r squared is smaller (0.5).
the relation between real gdp and republican votes is very strong (1.45) and the other two variables are significant with smaller effects

```python
quantile_model = smf.quantreg('OTHER_log ~ Q("unemp rate(log)") + Q("Poverty Percent, All Ages(log)") + Q("Real GDP(log)")', filtered_df)
quantile_fit = quantile_model.fit(q=0.5)
quantile_fit.summary()
```

this model has a low R squared, which means that other variables (different than economic indicators) explain better the results

```python
quantile_model = smf.quantreg('LIBERTARIAN_log ~ Q("unemp rate(log)") + Q("Poverty Percent, All Ages(log)") + Q("Real GDP(log)")', filtered_df)
quantile_fit = quantile_model.fit(q=0.5)
quantile_fit.summary()
```
